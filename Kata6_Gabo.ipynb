```python
# Lista de planetas

planetas = ['mercurio', 'venus', 'tierra', 'marte', 'jupiter', 'saturno', 'neptuno']
```

```python
# Solicitamos el nombre de un planeta *Pista:  input()*
solicitud = input('Ingrese el nombre del planeta (en minusculas)')
```

```python
# Busca el planeta en la lista

busqueda = planetas.index(solicitud)
```

```python
# Muestra los planetas más cercanos al sol

print('Estos son los planetas mas cercanos para ' + solicitud)
print(planetas[0:busqueda])
```

```python
# Muestra los planetas más lejanos al sol

print('Estos son los planetas mas lejanos ' + solicitud)
print(planetas[busqueda + 1:])
```
