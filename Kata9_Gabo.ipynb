```python
# Función para leer 3 tanques de combustible y muestre el promedio

def reporte(tanque_principal, tanque_externo, tanque_hidrogeno):
    promedio_total = (tanque_principal + tanque_externo + tanque_hidrogeno) / 3
    return f"""Reporte Completo:
    Promedio Total: {promedio_total}%
    Tanque Principal: {tanque_principal}%
    Tanque Secundario: {tanque_externo}%
    Tanque Hidrogeno: {tanque_hidrogeno}% 
    """
```

```python
# Llamamos a la función que genera el reporte print(funcion(tanque1, tanque2, tanque3))
print(reporte(80, 70, 85))
```

```python
# Función promedio 
def prom(values):
    total = sum(values)
    cantidad = len(values)
    return total / cantidad

prom([80,70,85])
```

```python
# Actualiza la función
def reporte(tanque_principal, tanque_externo, tanque_hidrogeno):
    return f"""Reporte Completo:
    Promedio Total: {prom([tanque_principal, tanque_externo, tanque_hidrogeno])}%
    Tanque Principal: {tanque_principal}%
    Tanque Secundario: {tanque_externo}%
    Tanque Hidrogeno: {tanque_hidrogeno}% 
    """


# Call the updated function again with different values
print(reporte(88, 76, 70))
```

```python
# Función con un informe preciso de la misión. Considera hora de prelanzamiento, tiempo de vuelo, destino, tanque externo y tanque interno

def reporte_mision(prelanzamiento, tiempo_vuelo, destino, tanque_externo, tanque_interno):
    return f"""
    Destino de la misión: {destino}
    Total tiempo de viaje: {prelanzamiento + tiempo_vuelo} minutos
    Combustible total restante: {tanque_externo + tanque_interno} litros
    """

print(reporte_mision(14, 51, "Luna", 200000, 300000))
```
