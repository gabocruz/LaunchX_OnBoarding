```python
# Ciclo for para recorrer la cadena

parrafo = """Interesting facts about the Moon. The Moon is Earth's only satellite. There are several interesting facts about the Moon and how it affects life here on Earth. 
On average, the Moon moves 4cm away from the Earth every year. This yearly drift is not significant enough to cause immediate effects on Earth. The highest daylight temperature of the Moon is 127 C."""

oraciones = parrafo.split('. ')

claves = ["average", "temperature", "distance"]

for busqueda in oraciones:
    for claves in claves:
        if claves in busqueda:
            print(busqueda)
            break
```

```python
# Ciclo para cambiar C a Celsius

parrafo = """Interesting facts about the Moon. The Moon is Earth's only satellite. There are several interesting facts about the Moon and how it affects life here on Earth. 
On average, the Moon moves 4cm away from the Earth every year. This yearly drift is not significant enough to cause immediate effects on Earth. The highest daylight temperature of the Moon is 127 C."""

oraciones = parrafo.split('. ')

claves = ["average", "temperature", "distance"]

for cambio in oraciones:
    for claves in claves:
        if claves in cambio:
            print(cambio.replace(' C', ' Celsius'))
            break
```
